```python
Dict = {1:"january", 2:"february", 3:"march", 4:"april", 5:"may", 6:"june", 7:"july", 8:"august", 9:"september", 10:"october", 11:"november", 12:"december"}
print(Dict)
```

```python
month_Column = []
for rr in data2["arrival_date"]:
    str = rr
    str2 = str.split('/')
    month_Column.append(Dict[int(str2[1])])
```

```python
len(month_Column)
```

```python
data2["month_Column"] = month_Column
```

```python
data2["month_Column"]
```

```python
data2["month_Column"].unique()
```

```python
season_names = []

for tt in data2["month_Column"]:
    if tt in ["december", "january", "february"]:
        season_names.append("winter")
    elif tt in ["march", "april", "may"]:
        season_names.append("spring")
    elif tt in ["june", "july", "august"]:
        season_names.append("summer")
    elif tt in ["september", "october", "november"]:
        season_names.append("autumn")

```

```python
data2["season_names"] = season_names
```

```python
data2.head()
```

```python
import pandas as pd 
df = pd.Timestamp('04/03/2019')
print(df.dayofweek)
```

```python
day_of_week = []
for rr in data2["arrival_date"]:
    str = rr
    df = pd.Timestamp(rr)
    day = df.dayofweek
    day_of_week.append(day)
```

```python
len(day_of_week)
```

```python
data2["day"] = day_of_week
```

```python
Q1 = np.percentile(data2['modal_price'],25,interpolation='midpoint')

Q3 = np.percentile(data2['modal_price'],75,interpolation='midpoint')
IQR = Q3-Q1
```

```python
upper = np.where(data2['modal_price'] >= (Q3+1.5*IQR))

lower = np.where(data2['modal_price'] <= (Q1-1.5*IQR))
```

```python
print(upper[0],lower[0])
```

```python
dist = (data2['commodity'])
distset = set(dist)
dd = list(distset)
dictofWords_commodity = { dd[i]:i for i in range(0,len(dd))}
data2['commodity'] = data2['commodity'].map(dictofWords_commodity)
```

```python
dist = (data2['state'])
distset = set(dist)
dd = list(distset)
dictofWords_state = { dd[i]:i for i in range(0,len(dd))}
data2['state'] = data2['state'].map(dictofWords_state)
```

```python
dist = (data2['district'])
distset = set(dist)
dd = list(distset)
dictofWords_district = { dd[i]:i for i in range(0,len(dd))}
data2['district'] = data2['district'].map(dictofWords_district)
```

```python
dist = (data2['market'])
distset = set(dist)
dd = list(distset)
dictofWords_market = { dd[i]:i for i in range(0,len(dd))}
data2['market'] = data2['market'].map(dictofWords_market)
```

```python
dist = (data2['month_Column'])
distset = set(dist)
dd = list(distset)
dictofWords_month = { dd[i]:i for i in range(0,len(dd))}
data2['month_Column'] = data2['month_Column'].map(dictofWords_month)
```

```python
dist = (data2['season_names'])
distset = set(dist)
dd = list(distset)
dictofWords_season = { dd[i]:i for i in range(0,len(dd))}
data2['season_names'] = data2['season_names'].map(dictofWords_season)
```

```python
features = data2[['commodity', 'state', 'district', 'market', 'month_Column', 'season_names', 'day']]

labels = data2['modal_price']
```

```python
from sklearn.model_selection import train_test_split
Xtrain,Xtest,Ytrain,Ytest = train_test_split(features,labels,test_size=0.2,random_state=2)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report
from sklearn import metrics
from sklearn import tree
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.datasets import make_regression
regr = RandomForestRegressor(max_depth=1000, random_state=0)
regr.fit(Xtrain, Ytrain)
```

```python
Xtest[0:1]
```

```python
y_pred = regr.predict(Xtest)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(Ytest, y_pred)
```
